# A. Write a program

```python
data1=(data['LoanStatus']=='Completed')| (data['LoanStatus']=='Defaulted')|\
                  (data['LoanStatus'] == 'Chargedoff')
data2=data[data1]

def new_c(category):
    if category['LoanStatus'] == 'Chargedoff':
        return 'Defaulted'
    else:
        return category['LoanStatus']
data2['LoanStatus'] = data2.apply(new_c, axis=1)
data2['LoanStatus'].value_counts()
```

there are 332 completed loans and 146 defaulted loans after categorising data

# C. Display the box plot for ProsperRating (Alpha) vs LoanOriginalAmount and hue = Loan Status[Completed,Defaulted]. Write your observations

```python
plt.figure(figsize = [10, 8])
sns.boxplot(data=data2, x='ProsperRating (Alpha)', y='LoanOriginalAmount', hue='LoanStatus')
plt.show()
```

Most of the defaulted loans with low Prosper rating

# B.write a program

```python
categories = {1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'}
def list_categories(cat):
    loan_category = cat['ListingCategory (numeric)']
    if  loan_category in categories:
        return categories[loan_category]
    else:
        return categories[7]
    
data['ListingCategory (numeric)'] = data.apply(list_categories, axis=1)
data['ListingCategory (numeric)'].value_counts()
```

# D. Display the catplot for ProsperRating (Alpha) vs ListingCategory (numeric)[Debt Consolidation,Other,Business,Home Improvement,Auto] and hue = Loan Status[Completed,Defaulted]. Write your observations.

```python
plt.figure(figsize=[10,8])
sns.catplot(x = 'ProsperRating (Alpha)', hue = 'LoanStatus', col = 'ListingCategory (numeric)',
              data = data, kind = 'count',col_wrap=3)
plt.show()
```

# no clear difference exists between prosper rating for defaulted loans 
debt consolidation category has highest number of loans issued in current loans than all other listing categories
